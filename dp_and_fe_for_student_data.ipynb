# **‡∏à‡∏±‡∏î‡∏Å‡∏≤‡∏£‡∏Å‡∏±‡∏ö‡∏Ñ‡∏ô‡∏ó‡∏µ‡πà‡∏≠‡∏≤‡∏¢‡∏∏‡πÄ‡∏õ‡πá‡∏ô NaN ‡πÉ‡∏´‡πâ‡πÄ‡∏õ‡πá‡∏ô ‡∏Ñ‡πà‡∏≤‡∏Å‡∏•‡∏≤‡∏á‡∏Ç‡∏≠‡∏á‡∏Ñ‡∏ô‡∏ó‡∏±‡πâ‡∏á‡∏´‡∏°‡∏î üìä**
# **‡∏à‡∏±‡∏î‡∏Å‡∏≤‡∏£‡∏Å‡∏±‡∏ö‡∏Ñ‡∏ô‡∏ó‡∏µ‡πà‡∏≠‡∏≤‡∏¢‡∏∏‡∏ú‡∏¥‡∏î‡∏õ‡∏Å‡∏ï‡∏¥** 100++ ‡∏õ‡∏µ**

```python
df_student['‡∏≠‡∏≤‡∏¢‡∏∏'].fillna(df_student['‡∏≠‡∏≤‡∏¢‡∏∏'].median(), inplace=True)

# ‡∏Ñ‡∏≥‡∏ô‡∏ß‡∏ì IQR (Interquartile Range)
Q1 = df_student['‡∏≠‡∏≤‡∏¢‡∏∏'].quantile(0.25)
Q3 = df_student['‡∏≠‡∏≤‡∏¢‡∏∏'].quantile(0.75)
IQR = Q3 - Q1

# ‡∏™‡∏£‡πâ‡∏≤‡∏á‡∏Ç‡∏≠‡∏ö‡πÄ‡∏Ç‡∏ï‡∏™‡∏≥‡∏´‡∏£‡∏±‡∏ö outlier
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# ‡∏Å‡∏£‡∏≠‡∏á‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏ó‡∏µ‡πà‡∏≠‡∏¢‡∏π‡πà‡∏ô‡∏≠‡∏Å‡∏ä‡πà‡∏ß‡∏á (‡πÄ‡∏≠‡∏≤‡πÄ‡∏â‡∏û‡∏≤‡∏∞‡∏ó‡∏µ‡πà‡πÑ‡∏°‡πà‡πÉ‡∏ä‡πà outlier)
df_student = df_student[(df_student['‡∏≠‡∏≤‡∏¢‡∏∏'] >= lower_bound) & (df_student['‡∏≠‡∏≤‡∏¢‡∏∏'] <= upper_bound)]

```

```python
df_student
```

# **‡∏ó‡∏≥‡πÉ‡∏´‡πâ‡πÄ‡∏û‡∏®‡πÅ‡∏•‡∏∞‡∏ä‡∏∑‡πà‡∏≠‡∏Ñ‡∏ì‡∏∞‡πÄ‡∏õ‡πá‡∏ô‡∏°‡∏≤‡∏ï‡∏£‡∏ê‡∏≤‡∏ôüöπüö∫&üìù**

```python
gender_mapping = {
    '‡∏ä': '‡∏ä‡∏≤‡∏¢',
    '‡∏ä‡∏≤‡∏¢': '‡∏ä‡∏≤‡∏¢',
    'M': '‡∏ä‡∏≤‡∏¢',
    'Male': '‡∏ä‡∏≤‡∏¢',

    '‡∏ç': '‡∏´‡∏ç‡∏¥‡∏á',
    '‡∏´‡∏ç‡∏¥‡∏á': '‡∏´‡∏ç‡∏¥‡∏á',
    'F': '‡∏´‡∏ç‡∏¥‡∏á',
    'Female': '‡∏´‡∏ç‡∏¥‡∏á'
}
df_student['‡πÄ‡∏û‡∏®'] = df_student['‡πÄ‡∏û‡∏®'].map(gender_mapping)
df_student['‡πÄ‡∏û‡∏®'].fillna(df_student['‡πÄ‡∏û‡∏®'].mode()[0], inplace=True)
```

```python
faculty_mapping = {
    '‡∏ß‡∏¥‡∏ó‡∏¢‡πå': '‡∏ß‡∏¥‡∏ó‡∏¢‡∏≤‡∏®‡∏≤‡∏™‡∏ï‡∏£‡πå',
    '‡∏Ñ‡∏ì‡∏∞‡∏ß‡∏¥‡∏ó‡∏¢‡∏≤‡∏®‡∏≤‡∏™‡∏ï‡∏£‡πå': '‡∏ß‡∏¥‡∏ó‡∏¢‡∏≤‡∏®‡∏≤‡∏™‡∏ï‡∏£‡πå',
    'Science': '‡∏ß‡∏¥‡∏ó‡∏¢‡∏≤‡∏®‡∏≤‡∏™‡∏ï‡∏£‡πå',
    'Sci': '‡∏ß‡∏¥‡∏ó‡∏¢‡∏≤‡∏®‡∏≤‡∏™‡∏ï‡∏£‡πå',
    '‡∏ß‡∏¥‡∏ó‡∏¢‡∏≤‡∏®‡∏≤‡∏™‡∏ï‡∏£‡πå': '‡∏ß‡∏¥‡∏ó‡∏¢‡∏≤‡∏®‡∏≤‡∏™‡∏ï‡∏£‡πå',

    '‡∏ß‡∏¥‡∏®‡∏ß‡∏∞': '‡∏ß‡∏¥‡∏®‡∏ß‡∏Å‡∏£‡∏£‡∏°‡∏®‡∏≤‡∏™‡∏ï‡∏£‡πå',
    '‡∏ß‡∏¥‡∏®‡∏ß‡∏Ø': '‡∏ß‡∏¥‡∏®‡∏ß‡∏Å‡∏£‡∏£‡∏°‡∏®‡∏≤‡∏™‡∏ï‡∏£‡πå',
    '‡∏ß‡∏¥‡∏®‡∏ß‡∏Å‡∏£‡∏£‡∏°': '‡∏ß‡∏¥‡∏®‡∏ß‡∏Å‡∏£‡∏£‡∏°‡∏®‡∏≤‡∏™‡∏ï‡∏£‡πå',
    'Engineering': '‡∏ß‡∏¥‡∏®‡∏ß‡∏Å‡∏£‡∏£‡∏°‡∏®‡∏≤‡∏™‡∏ï‡∏£‡πå',
    '‡∏ß‡∏¥‡∏®‡∏ß‡∏Å‡∏£‡∏£‡∏°‡∏®‡∏≤‡∏™‡∏ï‡∏£‡πå': '‡∏ß‡∏¥‡∏®‡∏ß‡∏Å‡∏£‡∏£‡∏°‡∏®‡∏≤‡∏™‡∏ï‡∏£‡πå',

    '‡∏®‡∏¥‡∏•‡∏õ‡πå': '‡∏®‡∏¥‡∏•‡∏õ‡∏®‡∏≤‡∏™‡∏ï‡∏£‡πå',
    'Arts': '‡∏®‡∏¥‡∏•‡∏õ‡∏®‡∏≤‡∏™‡∏ï‡∏£‡πå',
    'Liberal Arts': '‡∏®‡∏¥‡∏•‡∏õ‡∏®‡∏≤‡∏™‡∏ï‡∏£‡πå',
    '‡∏®‡∏¥‡∏•‡∏õ‡∏®‡∏≤‡∏™‡∏ï‡∏£‡πå': '‡∏®‡∏¥‡∏•‡∏õ‡∏®‡∏≤‡∏™‡∏ï‡∏£‡πå'
}
df_student['‡∏Ñ‡∏ì‡∏∞'] = df_student['‡∏Ñ‡∏ì‡∏∞'].map(faculty_mapping)


most_common_faculty = df_student['‡∏Ñ‡∏ì‡∏∞'].mode()[0]
df_student['‡∏Ñ‡∏ì‡∏∞'].fillna(most_common_faculty, inplace=True)
```

# ‡πÅ‡∏Å‡πâ‡πÑ‡∏Ç GPA ‡∏ú‡∏¥‡∏î‡∏õ‡∏Å‡∏ï‡∏¥‚úçÔ∏è

```python
df_student.loc[df_student['GPA'] < 0, 'GPA'] = np.nan
df_student.loc[df_student['GPA'] > 4, 'GPA'] = 4.0
mean_gpa = df_student['GPA'].mean()
df_student['GPA'].fillna(mean_gpa, inplace=True)
```

```python
df_student
```
